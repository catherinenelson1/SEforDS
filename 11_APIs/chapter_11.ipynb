# APIs

### Getting Data from the SDG API

```python
import requests
```

```python
response = requests.get('https://unstats.un.org/SDGAPI/v1/sdg/Goal/List')
```

```python
response.status_code
```

```python
response.headers['content-type']
```

```python
response.json()[4]
```

```python
response = requests.get('https://unstats.un.org/sdgapi/v1/sdg/Goal/5/Target/List?includechildren=true')
```

```python
response.status_code
```

```python
for i in response.json()[0]['targets']:
    print(i)
    print('')
```

### Calling your own API

```python
response = requests.get('http://127.0.0.1:8000/country_trendline/India')
```

```python
response.status_code
```

```python
response.json()
```

```python
url = 'http://127.0.0.1:8000/fit_trendline/'

json_data = {'timestamps': [2000, 2001, 2002],
           'data': [0.5, 0.6, 0.7]}
```

```python
response = requests.post(url=url, json=json_data)
```

```python
response.json()
```
